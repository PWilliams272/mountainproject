# Holcomb Valley Pinnacles Area Probe

This notebook does a very light test against Holcomb Valley Pinnacles. It fetches one area page, parses it with the package code, and shows which child areas are available for deeper scraping.

Recommended kernel: the workspace `.venv`.

```python
from pathlib import Path
import sys

CWD = Path.cwd().resolve()
PROJECT_ROOT = CWD if (CWD / 'src').exists() else CWD.parent
SRC = PROJECT_ROOT / 'src'
if SRC.exists() and str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

DATA_ROOT = PROJECT_ROOT / 'data'

PROJECT_ROOT
```

```python
from mountainproject.scraper.client import MountainProjectClient
from mountainproject.scraper.extract import parse_area_page
ROOT_AREA_URL = 'https://www.mountainproject.com/area/105805238/holcomb-valley-pinnacles'
CACHE_DIR = DATA_ROOT / 'notebook_cache' / 'area_probe'

client = MountainProjectClient(
    delay_seconds=0.0,
    cache_dir=CACHE_DIR,
    user_agent='Mozilla/5.0',
)
response = client.fetch_text(ROOT_AREA_URL)
area = parse_area_page(response.text, response.url)

(area.name, area.area_id, len(area.child_area_urls), len(area.route_urls))
```

```python
import pandas as pd

child_areas = pd.DataFrame({'child_area_url': area.child_area_urls})
child_areas['area_id'] = child_areas['child_area_url'].str.extract(r'/area/(\d+)/')
child_areas.head(20)
```
